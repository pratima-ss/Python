```python
%%writefile app.py
import streamlit as st
import pandas as pd
import numpy as np
import pickle
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn import preprocessing
with open('model_for_heart_disease_prediction' , 'rb') as f:
    lr = pickle.load(f)
primaryColor="#b8c1ec"
st.title("Heart disease predictor")
def get_user_input():
  age = st.number_input("Enter age")
  sex=st.slider( "Gender (0->Female 1->Male)",0, 1)
  chest_pain=st.slider("Chest Pain (0->Asymptomatic  1->Atypical angina 2->Non-anginal pain 3->Typical angina)",0,3)
  resting_blood_pressure = st.number_input("Enter resting blood pressure")
  cholestoral= st.number_input("Enter cholestoral level")
  fasting_blood_sugar=st.slider( "Select fasting blood sugar level (0->Greater than 120 mg/ml 1->Lower than 120 mg/ml)",0,1)
  rest_ecg=st.slider( "Select rest ecg status (0->Left ventricular hypertrophy 1->Normal 2->ST-T wave abnormality)",0,2)
  maximum_heart_rate= st.number_input("Enter maximum heart beat rate")
  exercise_induced_angina=st.slider("Exercise induced angina (0->No 1->Yes)",0,1)
  oldpeak=st.number_input("Enter old peak")
  slope=st.slider("Select slope (0->Downslopping 1->Upslopping 2->Flat)",0,2)
  vessels_colored_by_flourosopy=st.slider("Vessels colored by flouroscopy (0->Four 1->One 2->Three 3->Two 4->Zero)",0,4)
  thalassemia=st.slider("Thalassemia (0->Fixed defect 1->No 2->Normal 3->Reversible)",0,3)

  user_data={
    'age':age,
    'sex':sex,
    'chest_pain_type':chest_pain,
    'resting_blood_pressure':resting_blood_pressure,
    'cholestoral':cholestoral,
    'fasting_blood_sugar':fasting_blood_sugar,
    'rest_ecg':rest_ecg,
    'Max_heart_rate':maximum_heart_rate,
    'exercise_induced_angina':exercise_induced_angina,
    'oldpeak':oldpeak,
    'slope':slope,
    'vessels_colored_by_flourosopy':vessels_colored_by_flourosopy,
    'thalassemia':thalassemia,
    }
  features = pd.DataFrame(user_data, index = [0])
  return features
user_input = get_user_input()
input_data_as_numpy_array= np.asarray(user_input)
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)
print(input_data_reshaped)
prediction = lr.predict(input_data_reshaped)
print(prediction)

if (prediction[0]== 0):
  st.write('The Person does not have a Heart Disease')
  #print('The Person does not have a Heart Disease')
elif(prediction[0]==1):
  st.write('The Person has a Heart Disease')
  #print('The Person has Heart Disease')


 
```

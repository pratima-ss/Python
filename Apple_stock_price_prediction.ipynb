```python
import numpy as np
import pandas as pd
import plotly.graph_objects as go
import csv
import ast
import requests
from urllib.request import urlopen
import certifi
import json
```

```python
url="https://financialmodelingprep.com/api/v3/historical-price-full/AAPL?apikey=ee657a7b8f40314a49b07b41c1b9c3dd"
```

```python
response = requests.get(url).json()
```

```python
response
```

```python
response=response['historical']
df = pd.DataFrame(response)
df.drop(df.columns[[7,8,9,10,11,12]], axis = 1, inplace = True)
print(df.head())
```

```python
dataa=df
```

```python
figure=go.Figure(data=[go.Candlestick(x=dataa["date"],open=dataa["open"],high=dataa["high"],low=dataa["low"],close=dataa["close"])])
```

```python
figure.update_layout(title="Apple Stock Price Analysis",xaxis_rangeslider_visible=False)
figure.show()
```

```python
print(dataa.corr())
```

```python
from autots import AutoTS
```

```python
model= AutoTS(forecast_length=50,frequency='infer',ensemble='simple')
```

```python
model=model.fit(dataa,date_col='date',value_col='close',id_col=None)
```

```python
print(model)
```

```python
model.results()
```

```python
validation_results = model.results("validation")
```

```python
print(validation_results)
```

```python
prediction=model.predict()
```

```python
forecast=prediction.forecast
```
